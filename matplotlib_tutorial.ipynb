## Matplotlib Tutorial

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4,5],[25,0,50,65,100])
```

```python
x=[1,2,3,4,5]
y=[25,0,50,65,100]
fig=plt.figure()
ax=fig.add_axes([0,1,5,2])
ax.plot(x,y)
```

```python
# matplotolib workflow

#import lib
import matplotlib.pyplot as plt

# prepare data
x=[1,2,3,4,5]
y=[25,0,50,65,100]

# setup plot
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot()

#plot data
ax.plot(x,y)

# cutomize plot
ax.set(title="plot" ,xlabel="X",ylabel="Y")
```

```python
# scatter plot
x=np.linspace(0,10,100)
fig,ax=plt.subplots(figsize=(5,5))
ax.scatter(x,np.exp(x))
```

```python
# bar plot
fig,ax=plt.subplots(figsize=(5,5))
ax.bar(x[0:6],np.exp(x[0:6]))
```

```python
# histogram
fig,ax=plt.subplots(figsize=(4,4))
ax.hist(np.random.random(10))
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2 ,figsize=(5,5))
ax[0,0].plot(x[1:],np.log(x[1:]))
ax[0,1].scatter(np.random.rand(10),np.random.rand(10))
ax[1,0].bar(x[0:6],np.exp(x[0:6]))
ax[1,1].hist(np.random.random(10))
```
